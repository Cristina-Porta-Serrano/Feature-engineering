*Continua amb el conjunt de dades de tema esportiu que t'agradi i aplica l'anàlisi de components principals.*

L'anàlisi de components principals **(Principal Component Analysis - PCA)** és un mètode de reducció de dimensionalitat que permet simplificar la complexitat d'espais amb múltiples dimensions a la vegada que conserva la seva informació. Aquest mètode permet "condensar" la informació aportada per múltiples variables en només uns quants components. Dues de les principals aplicacions del PCA són la visualització i el preprocessat de predictors previ ajust de models supervisats.

Comencem aplicant un **PCA de 3 components** per veure quin percentatge de variància surt. Si cal, modificarem el número de PCA per ajustar a la variància desitjada.

```python
pca_features = ['Age', 'Height', 'Weight']
Z = hockey_scaled.loc[:, pca_features]
pca = PCA(n_components=3)
Z_pca = pca.fit_transform(Z)
pca_hockey = pd.DataFrame(data= Z_pca , columns = ['PCA1', 'PCA2', 'PCA3']) 
pca_hockey 
```

```python
pca.explained_variance_ratio_
```

```python
cumulative_var_ratio = np.cumsum(pca.explained_variance_ratio_)
cumulative_var_ratio
```

Veiem que el primer component principal explica un **63'9%** de la variació total de les dades originals, mentre que el segon explica un **30'8%**. Conjuntament, els dos components principals expliquen al voltant del 95% de la variació total, un percentatge bastant elevat, tot i que per arribar al **95%** necessitem un tercer component. 

```python
plt.rcParams["figure.figsize"] = (12,6)

fig, ax = plt.subplots()
x = np.arange(1, 4, 1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.plot(x, y, marker='o', linestyle='-', color='g')

plt.xlabel('Number of Components')
plt.xticks(x)
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='Lightgreen', linestyle='--')
plt.text(2, 0.92, '95% cut-off', color = 'g', fontsize=10)

ax.grid(axis='x')
plt.show()
```

```python
pca = PCA(.95)
Z_pca = pca.fit_transform(Z)
pca.n_components_
```

Entre la gràfica i el càlcul veiem que, per tenir un percentatge de variança del 95%, hem de tenir com a mínim 3 components (PCA), ja que amb 2 ens quedem a 0'947. Aquest és el valor que maximitza la variança entre la nostra raw data i el dataset projectat.

```python
plt.bar(range(0,len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_, alpha=0.5, align='center', label='Individual explained variance', color="Mediumseagreen")
plt.step(range(0,len(cumulative_var_ratio)), cumulative_var_ratio, where='mid',label='Cumulative explained variance', color="Seagreen")
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.axhline(y=0.95, color='Lightgreen', linestyle='--')
plt.text(2, 0.92, '95% cut-off', color = 'g', fontsize=10)
plt.show()
```

En aquesta gràfica tenim les barres, utilitzades per representar les variances explicades individuals, i la gràfica de "steps", que representa la variància explicada acumulada dels diferents components principals. Veiem que els índex dels components principals comencen a 0 i acaben en 2 però realment és de 1 a 3. Tornem a veure que amb el component principal 2 ens quedem quasi al 95% de variància explicada acumulada però que per arribar al 95% real necessitem un tercer component principal.

```python
pca = PCA()
components = pca.fit_transform(hockey_scaled[pca_features])
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(components, labels=labels, dimensions=range(3), color=hockey_scaled["Age"])
fig.update_traces(diagonal_visible=False)
fig.show()
```

```python
Q1 = hockey.quantile(0.25)
Q3 = hockey.quantile(0.75)
IQR = Q3 - Q1

outliers = ((hockey < (Q1 - 1.5 * IQR)) | (hockey > (Q3 + 1.5 * IQR))).sum()
outliers
```

Veiem que tenim pocs outliers en les nostres dades. Al haver-hi tant pocs, podriem esborrar-los però farem una normalització de les dades amb Robust Scaler tenint-los en compte per reduir el seu impacte.

Aqusta tècnica escala característiques utilizant estadístiques que son robustes per a valors atípics. Aquest mètode elimina la mediana i escala les dades en el rang entre el 1er quartil i el 3er quartil, és a dir, en el rang interquartílic. S'emmatatzemen la mediana i el rang interquartílic per a que es puguin utilitzar en dades futures mitjançant el mètode de transformació. Si hi ha valors atípics en el conjunt de dades, la mediana i el rang intercuartílic proporcionen millors resultats i superen la mitja i la variança de la mostra. RobustScaler utilitza el rang interquartílic per a que sigui robust a valors atípics.
